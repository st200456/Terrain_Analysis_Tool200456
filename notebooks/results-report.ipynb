## 1. Results
Our Sample AOI is a 92 km² area around the River Neckar in the "Remseck am Neckar - Bad-Canstatt" Region, defined by:

- west: 9.16362763394136
- south: 48.78900873335118
- east : 9.28804778738413
- north : 48.878862334264085

The terrain includes moderate elevation variation and typical central European land cover patterns.

### - Terrain Roughness and DEM

```python
import matplotlib.pyplot as plt
import rasterio

def plot_raster(path, title, cmap="terrain"):
    with rasterio.open(path) as src:
        data = src.read(1)
        plt.figure(figsize=(8,6))
        plt.imshow(data, cmap=cmap)
        plt.colorbar()
        plt.title(title)
        plt.show()

plot_raster("../results/roughness.tif", "Terrain Roughness", cmap="viridis")
plot_raster("../results/dem_wgs84.tif", "DEM - WGS84")
plot_raster("../results/LandCover_matched.tif", "Landcover")
```

### - Thalweg

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show

with rasterio.open("../results/dem_wgs84_utm32n.tif") as src:
    fig, ax = plt.subplots(figsize=(8,6))
    show( src , ax=ax , cmap="terrain")

gdf = gpd.read_file("../results/Thalweg.shp")
gdf.plot(ax=plt.gca(), color="red", linewidth=2)
plt.title("Thalweg over DEM")
plt.show()
```
